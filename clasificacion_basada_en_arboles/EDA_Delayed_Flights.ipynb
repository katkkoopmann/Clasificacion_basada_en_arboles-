```python
#Manejar valores faltantes con scikit
from sklearn.impute import SimpleImputer
import numpy as np
#Seleccionamos valores numericos
numeric_st_mean = df.select_dtypes(include=['int64', 'float64']).columns
#Eliminamos las variables en formato HHMM, ya que la estrategia mean no estaria siendo interpretable en estos casos.
numeric_st_mean = numeric_st_mean.drop(['ArrTime', 'CRSArrTime', 'DepTime', 'CRSDepTime'])
#Aplicamos simple imputer con estrategia de reemplazar con la media al resto de variables
imputer = SimpleImputer(strategy='mean')
df[numeric_st_mean] = imputer.fit_transform(df[numeric_st_mean])
#Separamos las variables en formato hora para imputar de diferente forma
hh_mm_columns = ['ArrTime', 'CRSArrTime', 'DepTime', 'CRSDepTime']
#Como no tenemos la informacion de vuelo es preferible rellenar con 0, para no cambiar la prediccion.
imputer_hh_mm = SimpleImputer(strategy='constant', fill_value = 0)
df[hh_mm_columns] = imputer.fit_transform(df[hh_mm_columns])
#Seleccionamos las variables categoricas.
categorical_columns = df.select_dtypes(include=['object']).columns
#Rellenamos los valores nulos con una nueva categoria que seria desconocido.
imputer = SimpleImputer(strategy='constant', fill_value='Desconocido')
df[categorical_columns] = imputer.fit_transform(df[categorical_columns])
```

```python
import pandas as pd
import numpy as np

# Seleccionar columnas numéricas
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Inicializar un contador total de outliers
total_outliers = 0

# Detectar outliers para cada columna numérica
outlier_summary = {}

for column in numeric_columns:
    if column in ['Year']:  # Excluir columnas irrelevantes si es necesario
        continue

    # Calcular cuartiles y rango intercuartílico (IQR)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Definir límites para detectar outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identificar valores fuera de los límites
    outliers = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()

    # Actualizar total de outliers
    total_outliers += outliers

    # Guardar detalles en el resumen
    outlier_summary[column] = outliers

# Crear un DataFrame resumen de outliers
outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outliers'])
outlier_df = outlier_df.sort_values(by='Outliers', ascending=False)

# Mostrar resultados
print("Número total de outliers en el dataset:", total_outliers)
print("\nResumen de outliers por columna:")
print(outlier_df)

```

```python
'''
Los outliers en retrasos de vuelo pueden afectar los modelos predictivos porque son eventos raros que no reflejan la norma.
Si no se gestionan, pueden distorsionar los resultados y llevar a conclusiones incorrectas sobre las causas de los retrasos o la gestion de los vuelos.
'''
```

```python
#Imputar outliers con los limites
import pandas as pd
import numpy as np

# Seleccionar columnas numericas
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

for column in numeric_columns:
    #Calcular cuartiles y rango intercuartil
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    #Definir limites para detectar outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    #Imputar outliers con los limites
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

print(df[numeric_columns].describe())
```
