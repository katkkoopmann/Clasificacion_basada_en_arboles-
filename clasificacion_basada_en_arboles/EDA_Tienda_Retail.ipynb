```python
#Creamos nuevas columnas desde la variable fecha para utilizar mejor en la prediccion
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Day'] = pd.DatetimeIndex(df['Date']).day
df
```

```python
#Eliminamos la columna fecha
df = df.drop('Date', axis=1)
df
```

```python
#Revisamos si hay valores nulos
df.isnull().sum()
#Como no hay valores nulos no hay cambios que realizar.
```

```python
#Revisamos si hay outliers
import pandas as pd
import numpy as np
#Seleccionamos las columnas numericas
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

#Inicializamos un contador total de outliers
total_outliers = 0

#Detectar outliers para cada columna numerica
outlier_summary = {}

for column in numeric_columns:

    #Calcular cuartiles y rango intercuartilico
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    #Definir limites para detectar outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    #Identificar valores fuera de los limites
    outliers = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()

    #Actualizar total de outliers
    total_outliers += outliers

    #Guardar detalles en el resumen
    outlier_summary[column] = outliers

#Crear un DataFrame resumen de outliers
outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outliers'])
outlier_df = outlier_df.sort_values(by='Outliers', ascending=False)

#Mostrar resultados
print("Número total de outliers en el dataset:", total_outliers)
print("\nResumen de outliers por columna:")
print(outlier_df)
```

Graficos para entender la distribucion de variables

```python
#Histogramas para visualizar distribucion
import matplotlib.pyplot as plt
import seaborn as sns
numeric_columns = df[['Age', 'Total Amount']]
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, kde=True)
    plt.title(f'Histograma de {column}')
    plt.show()
```

```python
#Grafico de barras para ver la frecuencia de aparicion
import matplotlib.pyplot as plt
import seaborn as sns
bar_columns = df[['Gender', 'Product Category', 'Month', 'Day', 'Quantity', 'Price per Unit']]
for column in bar_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column)
    plt.title(f'Frecuencia de {column}')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Diagrama de dispersion de Age vs Total Amount para ver si hay relacion
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Total Amount')
plt.title('Diagrama de Dispersión de Age vs Total Amount')
plt.show()
```

```python
#Grafico de barras apilado para ver la proporcion de compra de categorias por genero
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Product Category', hue='Gender', palette='Set2')
plt.title('Proporción de Compra de Categorías por Género')
plt.xlabel('Categoría de Producto')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.legend(title='Género')
plt.show()
```

```python
#Mapa de calor para mostrar la correlacion entre variables
import matplotlib.pyplot as plt
import seaborn as sns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de Correlación')
```
